### 1.1 Simulación

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
agua = [15 + x + np.random.normal() for x in range(1, 90)]
poblacion = [20 + 2*x + np.random.normal()*10 for x in range(1, 90)] 
```

```python
df = pd.DataFrame(zip(agua, poblacion), columns=['Consumo de agua','Población'])
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Consumo de agua', y='Población')
```

```python
horas_estudio = [9 + x + np.random.normal()*2 for x in range(1, 20)]
alumnos_reprobados = [20 - x + np.random.normal()*4 for x in range(1,20)] 
```

```python
df = pd.DataFrame(zip(horas_estudio, alumnos_reprobados), columns=['Horas de estudio','Alumnos reprobados'])
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Horas de estudio', y='Alumnos reprobados')
```

### 1.2 A/B test

```python
tasa_a = [0.15 + np.random.normal()/10 for x in range(1, 35)]
tasa_b = [0.15 + np.random.normal()/10 for x in range(1, 35)]
```

```python
df = pd.DataFrame(zip(tasa_a, tasa_b), columns=['Tasa_alg_A','Tasa_alg_B'])
```

```python
plt.figure(figsize=(15, 6))
sns.histplot(data = df, color="navy", kde=True)
```

> Diferencia de médias, es la mejor alternativa para medir el rendimiento de la tasa de conversión diario entre los dos algoritmos. 

H_0:  $μ_a = μ_b$

H_1:  $μ_a > μ_b$

$Z_{cal} = \frac{(\bar{x_A}-\bar{x_B})-(\bar{\mu_A}-\bar{\mu_B})}{\sqrt{\frac{s^2_A}{n_A} +\frac{s^2_B}{n_B}  }}$
